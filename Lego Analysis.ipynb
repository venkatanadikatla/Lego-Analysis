```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
parent = parent.rename(columns={'name':'parent_theme'}).copy()
parent.tail()
```

```python
df.head()
```

```python
merge = df.merge(parent, on ='parent_theme')
merge.head()
```

```python
merge['set_num'].isnull().sum()
```

```python
licensed = merge[merge['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
# you don't have to call out a boolean value as there the above row itself a boolean value
licensed.head()
```

```python
starwars = licensed[licensed['parent_theme']=='Star Wars']
starwars.head()
```

```python
starwars.shape
```

```python
the_force = int(starwars.shape[0]/licensed.shape[0]*100)
print(the_force, '%')
```
